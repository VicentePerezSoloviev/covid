```python
import pandas as pd

dt = pd.read_csv('dataset.csv')

dt['datetime'] = pd.to_datetime(dt['dateRep'], format='%d/%m/%Y')
dt = dt.set_index('datetime')

dt = dt.iloc[::-1]

start_date = '03/2020'
dt = dt[start_date:]
dt = dt[dt['countriesAndTerritories'] == 'Spain']

for i in ['cases', 'deaths']:
    dt[i] = dt[i].abs()

casos = dt['cases']
muertes = dt['deaths']

from statsmodels.tsa.statespace.sarimax import SARIMAX

s_mod = SARIMAX(casos,
                order=(0, 1, 0),
                seasonal_order=(1, 0, 0, 12))

case_predictions = s_mod.fit().predict()
print(case_predictions.head())
casos.append(case_predictions[:5]).plot()
```
